```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score, precision_score, classification_report, roc_curve, auc
```

```python
df = pd.read_csv("data.csv")
df.isnull().any()
```

Видно, что в столбце `Unnamed: 32 ` есть пропущенные значения

```python
df["Unnamed: 32"].describe()
```

В этой переменной нет вообще ни одного значения... Дропаем её тогда!

```python
df = df.drop("Unnamed: 32", axis=1)
```

```python
df.info()
```

```python
df.diagnosis = df.diagnosis.map({"M":0, "B":1})
```

Думаю, что нет большого смысла строить зависимость между категотиями 'mean', 'se', 'worst' нет, построю отдельно для каждой

```python
sns.pairplot(df.iloc[:,2:11], corner = True, diag_kind="kde", kind="reg")
```

```python
sns.pairplot(df.iloc[:,12:21], corner = True, diag_kind="kde", kind="reg")
```

```python
sns.pairplot(df.iloc[:,22:31], corner = True, diag_kind="kde", kind="reg")
```

Видно, что у некоторых переменных явная зависимость, например, периметра от радиуса, что логично. Род зависимости многих переменных не очевиден. В принципе распределение у них всех нормальное.

```python
X = df.drop(["id", "diagnosis"], axis = 1).values
y = df["diagnosis"].values

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
clf = LogisticRegression(random_state=0, max_iter=2000).fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
recall_score(y_test, y_pred, average='macro')
```

```python
precision_score(y_test, y_pred, average='macro')
```

```python
print(classification_report(y_test, y_pred))
```

Какая красота! Значения статистик близки к 1, всё супер!

```python
def get_roc_auc(estimator, y_test):
    probs = estimator.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)

    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
get_roc_auc(clf, y_test)
```

Ну очень хороший классификатор!
